```python
from scipy.stats import f_oneway
```

```python

target_column = 'G3'  
f_dict = {}
p_dict = {}
for column in math_df[variable_dict["cat"]].columns:
    categories = math_df[column].unique()
    values = [math_df[math_df[column] == category][target_column] for category in categories]
    f_val, p_val = f_oneway(*values)
    f_dict[column] = f_val
    p_dict[column] = p_val


        
```

```python
  
def elbow_method(df, K, metrics ="euclidean", r= 42, model = None, cat_col = None): 
        distortions = []
        inertias = []
        mapping1 = {}
        mapping2 = {}
        cost = {}
        for k in range(1,K):
            # Building and fitting the model
            
            if model == KMeans :
                kmeanModel = model(n_clusters=k, random_state=r).fit(df)
                kmeanModel.fit(df)
                distortions.append(sum(np.min(cdist(df, kmeanModel.cluster_centers_,
                                                'euclidean'), axis=1)) / df.shape[0])
                inertias.append(kmeanModel.inertia_)
            
                mapping1[k] = sum(np.min(cdist(df, kmeanModel.cluster_centers_,
                                            'euclidean'), axis=1)) / df.shape[0]
                mapping2[k] = kmeanModel.inertia_
                
            if model == KPrototypes :
                kmeanModel = model(n_clusters=k, random_state=r).fit(df, categorical =cat_col) 
                cost[k] = kmeanModel.cost_

        if model == KMeans :
            return K, distortions, inertias
        if model == KPrototypes :
            print(K, cost)            
            return K, cost
```

```python
import matplotlib.pyplot as plt

def scree_plot(k_list, inertia_list):
    """
    Trace le scree plot pour une série de k-means
    
    Args:
        k_list: une liste des valeurs de k utilisées
        inertia_list: une liste des inerties correspondantes
        
    Returns:
        None (affiche le scree plot)
    """
    # Tracer le scree plot
    plt.plot(k_list, inertia_list, 'ro-', linewidth=2)
    plt.title("Scree Plot pour une série de clusters")
    plt.xlabel("Nombre de clusters")
    plt.ylabel("Score")
    plt.show()
```

```python

def grid_search(X, y):
    grid_dict ={}
    models_and_parameters = {
        'SVC': 
        {
            'model': SVC(),
            'params': 
            {
                'C': [1, 10, 100], 
                'kernel': ['linear', 'rbf']
            }
        },
        
        'RandomForestClassifier': 
        {
            'model': RandomForestClassifier(),
            'params': 
            {
                'n_estimators': [1, 5, 10],
                'max_depth': [5, 8, 15, 25, 30], 
                'min_samples_split': [2, 5, 10, 15, 100],
                'min_samples_leaf': [1, 2, 5, 10] 
            }
        },

        'LogisticRegression':
        {
            'model': LogisticRegression(),
            'params':
            {
                'C': [0.1, 1, 10, 100],
                'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
            }
        },

        'XGBoost':
        {
            'model': XGBClassifier(eval_metric='logloss'),
            'params':
            {
                'n_estimators': [50, 100, 150, 200],
                'learning_rate': [0.01, 0.1, 0.2, 0.3],
                'max_depth': [3, 5, 7, 9]
            }
        }
    }


    for model_name, mp in models_and_parameters.items():
        clf = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
        clf.fit(X, y)
        grid_dict[model_name] = [clf.best_params_, clf.best_score_]

    return grid_dict
```

```python
grid_dict = grid_search(X_train, y_train)
```
